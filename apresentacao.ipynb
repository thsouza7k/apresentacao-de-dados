```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
base = pd.read_csv('train.csv')
base.head()
```

```python
base.info()
# presen√ßa de valores nulos na coluna 'Postal Code'
```

```python
# Verificando os registros com Postal code vazio
base[base['Postal Code'].isnull()]
```

```python
# Verificando se existe outros registro com City == 'Burlington' e se possuem Postal Code 
base[(base.City == 'Burlington') & (base.State == 'Vermont') & (base['Postal Code']).notnull()] 
```

```python
# Atribuindo o postal code = 05401 
base.loc[(base.City == 'Burlington') & (base.State == 'Vermont') & (base['Postal Code']).isnull(), "Postal Code"] = 5401
base.loc[(base.City == 'Burlington') & (base.State == 'Vermont')]
```

```python
# Tratando as colunas de datas que est√£o com Dtype = object 
import datetime as dt
base["Order Date"] = pd.to_datetime(base["Order Date"], format="%d/%m/%Y")
base["Ship Date"] = pd.to_datetime(base["Ship Date"], format="%d/%m/%Y")
```

```python
base["Order Date"].min()
```

```python
base["Order Date"].max()
```

```python
# Criando uma coluna para separar o ano
base['Ano'] = base['Order Date'].dt.year
```

###  Visualizando vendas por ano

```python
base.groupby("Ano")["Sales"].sum()
```

```python
fig, ax = plt.subplots(figsize=(9,6))

soma_ano = base.groupby("Ano")["Sales"].sum()

ax.bar(soma_ano.index , soma_ano.values, linewidth=2, color="#84ba5b");
ax.set_title("VENDAS POR ANO",
             fontsize= 14,
             fontweight= "bold",
             color= "#84ba5b");
ax.xaxis.set_ticks([2015,2016,2017,2018]);
ax.tick_params(axis='x', labelsize=12)

for i in range(len(soma_ano.index)):
            ax.annotate('{: ,.0f}'.format(soma_ano.values[i]), (soma_ano.index[i], soma_ano.values[i]),ha="center", xytext=(0,7), textcoords="offset points", fontweight="bold")
ax.yaxis.set_visible(False)
ax.spines.left.set_visible(False)
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)


plt.show()
```

###  Visualizando vendas por m√™s

```python
# criando coluna m√™s
base["Mes"] = base["Order Date"].dt.month
base.head(2)
```

```python
soma_mes = base.groupby(["Mes", "Ano"])["Sales"].sum()
soma_mes.head()
```

```python
# Transformando os indices em colunas
soma_mes = soma_mes.reset_index()
soma_mes.head()
```

```python
fig, ax = plt.subplots(figsize=(16,4))
w = 0.2
ax.bar(soma_mes[soma_mes.Ano == 2015].Mes.values -w-w/2, soma_mes[soma_mes.Ano == 2015].Sales.values,
                                color="#00c698", label='2015',width=w )
ax.bar(soma_mes[soma_mes.Ano == 2016].Mes.values-w/2 , soma_mes[soma_mes.Ano == 2016].Sales.values,
                                color='#00a9b5', label='2016',width=w)
ax.bar(soma_mes[soma_mes.Ano == 2017].Mes.values+w/2 , soma_mes[soma_mes.Ano == 2017].Sales.values,
                                color="#008ac5", label='2017',width=w )
ax.bar(soma_mes[soma_mes.Ano == 2018].Mes.values+w+w/2 , soma_mes[soma_mes.Ano == 2018].Sales.values,
                                color="#0069c0", label="2018",width=w )
ax.legend()
for i in range(0,12):
            ax.annotate('{: ,.0f}k'.format(soma_mes[soma_mes.Ano == 2018].Sales.values[i]/1000), 
                        (soma_mes[soma_mes.Ano == 2018].Mes.values[i]+w+w/2, soma_mes[soma_mes.Ano == 2018].Sales.values[i]),
                        ha="center", 
                        xytext=(3,5), 
                        textcoords="offset points", 
                        color="#0069c0",
                        fontweight="bold")
ax.set_yticks(np.array([0, 100000]))
ax.yaxis.set_visible(False)
ax.spines.top.set_visible(False)
ax.spines.right.set_visible(False)
ax.spines.left.set_visible(False) 
ax.xaxis.set_ticks(np.arange(1,13))
ax.set_xticklabels(['Jan','Fev','Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'])
plt.show()
```

### Agrupando por item e ano

```python
itens_ano=base.groupby(["Product Name", "Ano"])["Sales"].sum()
itens_ano = itens_ano.reset_index()
itens_ano
```

```python
df_datas = pd.DataFrame([2018,2017,2016,2015])
df_datas.columns = ['Ano']
df_datas
```

```python
df_datas['uniao'] = 'unir'
itens['uniao'] = 'unir'
```

```python
top_n_itens
```

```python
produto_ano = pd.merge(
    top_n_itens,
    df_datas,
    how='outer',
    on="uniao"
)
produto_ano.head()
```

```python
itens_ano_grafico = pd.merge(
    produto_ano,
    itens_ano,
    how='left',
    on=['Product Name', 'Ano']
)
itens_ano_grafico.head()
```

```python
fig, ax =plt.subplots(figsize=(10,9))
#itens_ano_grafico[itens_ano_grafico.Ano == 2018]['Product Name'].values
# itens_ano_grafico[itens_ano_grafico.Ano == 2018].Sales_y.values - vendas ano a ano de cade item
hgt = 0.2
qtd_itens = len(top_n_itens)
ax.barh(np.arange(0, qtd_itens)-hgt-hgt/2, #valor de y
            itens_ano_grafico[itens_ano_grafico.Ano == 2018].Sales_y.values, #valor de x
            align='center', height=hgt,label='2018',
            color='#0059c0'
        )
ax.barh(np.arange(0, qtd_itens)-hgt/2, #valor de y
            itens_ano_grafico[itens_ano_grafico.Ano == 2017].Sales_y.values, #valor de x
            align='center', height=hgt,label='2017',
             color='#008ac5'
        )
ax.barh(np.arange(0, qtd_itens)+hgt/2, #valor de y
            itens_ano_grafico[itens_ano_grafico.Ano == 2016].Sales_y.values, #valor de x
            align='center', height=hgt,label='2016',
            color='#00a9b5'
        )
ax.barh(np.arange(0, qtd_itens)+hgt+hgt/2, #valor de y
            itens_ano_grafico[itens_ano_grafico.Ano == 2015].Sales_y.values, #valor de x
            align='center', height=hgt,label='2015',
             color='#00c698'
        )

for i in range(0,qtd_itens):
            ax.annotate('{: ,.1f}k'.format(itens_ano_grafico[itens_ano_grafico.Ano == 2018].Sales_y.values[i]/1000), 
                        (itens_ano_grafico[itens_ano_grafico.Ano == 2018].Sales_y.values[i], i - hgt-hgt/2 ),
                        ha="center", 
                        va="top",
                        xytext=(20,5), 
                        textcoords="offset points", 
                        color="#0059c0",
                        fontweight="bold")
            
ax.yaxis.set_ticks(np.arange(0,qtd_itens))
ax.set_yticklabels(top_n_itens["Product Name"])
ax.tick_params(axis='y', labelsize=10)
ax.invert_yaxis()
#ax.set_xticks(np.arange(0,45000, 5000))
ax.xaxis.set_visible(False)
ax.spines.top.set_visible(False)
ax.spines.right.set_visible(False) 
ax.spines.bottom.set_visible(False) 

plt.legend()
plt.show()
```
